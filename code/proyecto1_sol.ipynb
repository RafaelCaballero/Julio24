1) Descargar localmente el fichero

https://www.imf.org/-/media/Files/Publications/WEO/2022/April/English/text.ashx
Grabarlo con el nombre focus.pdf



```python
import requests

url = "https://www.imf.org/-/media/Files/Publications/WEO/2022/April/English/text.ashx"
nombre = "focus.pdf"
r = requests.get(url) # el fichero queda en la variable r

if r.status_code == 200:
  with open(nombre, 'wb') as f:
    f.write(r.content) # ahora lo grabamos localmente
else:
  print(r.status_code)

```

2) Descargar la tabla que ocupa la posición 7ª (notación [6], recordar que la primera es la [0] ) de la tabla

https://www.expansion.com/mercados/cotizaciones/indices/ibex35_I.IB.html

Tras cargar esta tabla en un dataframe df de pandas:

1)	Mostrarla con un gradiente de colores para ver los valores más altos



```python
import pandas as pd
url = "https://www.expansion.com/mercados/cotizaciones/indices/ibex35_I.IB.html"
df_list = pd.read_html(url,encoding="utf-8")
df = df_list[6]
display(df)
display(df.style.background_gradient(cmap='YlOrRd'))
```

Si se quieren ver todas las tablas...

```python
from IPython.display import display
for i in range(len(df_list)):
    print(i)
    display(df_list[i])
```
