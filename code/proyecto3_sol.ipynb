### Ejercicio 9
Crear una nueva columna "inc" que tome el valor "positivo" si Open-Close>0, "negativo" si Open-Close<0 e "igual" en otro caso.

Ayuda:

1.- Crear una la columna "inc" con uno de los valores, por ejemplo "igual", para todos los valores

2.- Ahora hacer filtros para los otros dos valores y utilizar la siguiente sintaxis

    df_google.loc[filtro1,"inc"] = "loquesea"
donde `filtro1`es el filtro, y "loquesea" el valor que se quiere cambiar

```python
# Solución
df_google["inc"] = "igual"
filtroPos = df_google.dif>0
filtroNeg = df_google.dif<0
df_google.loc[filtroPos,"inc"] = "positivo"
df_google.loc[filtroNeg,"inc"] = "negativo"
df_google
```

```python
# solución
df_google[filtroPos].to_excel("positivo.xlsx")
df_google[filtroNeg].to_excel("negativo.xlsx")

```
